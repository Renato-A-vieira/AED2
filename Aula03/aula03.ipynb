```python
def busca_lista (vetor, chave):
    n = len(vetor)
    for i in range(n):
        if vetor [i] == chave:
            return i
    return -1

lista = [1,2,3,4,5]
busca_lista (lista, 1)
```

```python
def busca_cliente (vetor, chave):
    n = len(vetor)
    for i in range(n):
        if vetor [i] == chave:
            return i
    return -1

cliente = [
    "renato", "claudio", "emanuel", "humberto", "carlos", "claudia", "sabrina", "andre"]
cliente_1 = "claudio"
cliente_2 = "amanda"

busca1 = busca_cliente(cliente, cliente_1)
if busca1 == 1:
    print(f" O cliente {cliente_1} foi encontrado na posicão {busca1}")
else:
    print(f"o {cliente_1} não foi encontrado")

busca2 = busca_cliente(cliente, cliente_2)
if busca2 == 1:
    print(f" O cliente {cliente_2} foi encontrado na posicão {busca2}")
else:
    print(f" O cliente {cliente_2} não foi encontrado")
```
