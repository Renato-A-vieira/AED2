2.(TEÓRICA): Por que o algoritmo de busca binária exige que o vetor esteja ordenado previamente? O que
poderia acontecer se tentássemos aplicar a busca binária em um conjunto de dados não ordenado?


pois por meio da ordenação conseguimos aplicar a logica correta da busca binaria, sem isso o elemento do meio nao passa a informacao confiavel sobre a posicao do alvo, sendo assim nao conseguiriamos ver se o elemento da esquerda ou direita é maior ou menor e entao descartar metade dos dados de maneira errada.

```python
#3 
def busca_binaria_recursiva(vetor, alvo, esq=0, dire=None):
    if dire is None:
        dire = len(vetor) -1
    if esq > dire:
        return -1
    meio = (esq + dire) // 2
    if vetor[meio] == alvo:
        return meio
    elif vetor[meio] < alvo:
        return busca_binaria_recursiva(vetor,alvo,meio + 1, dire)
    else:
        return busca_binaria_recursiva(vetor,alvo,esq,meio - 1)
```

```python
#4
def busca_sequencial(lista, alvo):
    comparacoes = 0
    for i in range(len(lista)):
        comparacoes += 1
        if lista[i] == alvo:
            return i, comparacoes
    return -1, comparacoes

def busca_binaria(lista, alvo):
    comparacoes = 0
    inicio = 0
    fim = len(lista) - 1
    while inicio <= fim:
        comparacoes += 1
        meio = (inicio + fim) // 2
        comparacoes += 1
        if lista[meio] == alvo:
            return meio, comparacoes
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1
    return -1, comparacoes

lista = []
for i in range(1000):
    lista.append(i)

alvo = 1000

indice_seq, comp_seq = busca_sequencial(lista, alvo)
indice_bin, comp_bin = busca_binaria(lista, alvo)

print("Busca Sequencial:")
print("  Índice:", indice_seq)
print("  Comparações:", comp_seq)
print("Busca Binária:")
print("  Índice:", indice_bin)
print("  Comparações:", comp_bin)
```

```python
def busca_binaria_produto(catalogo, nome_buscado):
    inicio = 0
    fim = len(catalogo) - 1
    
    while inicio <= fim:
        meio = (inicio + fim) // 2
        nome_meio = catalogo[meio]["nome"]
        
        if nome_meio == nome_buscado:
            return catalogo[meio]
        elif nome_meio < nome_buscado:
            inicio = meio + 1
        else:
            fim = meio - 1
    
    return None

catalogo = [
    {"nome": "celular", "preco": 1500, "descricao": "Smartphone 128GB"},
    {"nome": "fone", "preco": 200, "descricao": "Fone Bluetooth"},
    {"nome": "notebook", "preco": 3500, "descricao": "Notebook 8GB RAM"},
    {"nome": "tablet", "preco": 1200, "descricao": "Tablet 10 polegadas"},
    {"nome": "tv", "preco": 2500, "descricao": "TV 4K 50 polegadas"}
]

nome = input("Digite o nome do produto: ")

resultado = busca_binaria_produto(catalogo, nome)

if resultado:
    print(f"Produto: {resultado['nome']}")
    print(f"Preço: R$ {resultado['preco']}")
    print(f"Descrição: {resultado['descricao']}")
else:
    print("Produto não encontrado")
```

```python
def busca_binaria_livro(biblioteca, titulo_buscado):
    inicio = 0
    fim = len(biblioteca) - 1
    
    while inicio <= fim:
        meio = (inicio + fim) // 2
        titulo_meio = biblioteca[meio]["titulo"]
        
        if titulo_meio == titulo_buscado:
            return biblioteca[meio]
        elif titulo_meio < titulo_buscado:
            inicio = meio + 1
        else:
            fim = meio - 1
    
    return None

biblioteca = [
    {"titulo": "Dom Casmurro", "autor": "Machado de Assis", "ano": 1899},
    {"titulo": "Memorias Postumas", "autor": "Machado de Assis", "ano": 1881},
    {"titulo": "O Alienista", "autor": "Machado de Assis", "ano": 1882},
    {"titulo": "O Guarani", "autor": "Jose de Alencar", "ano": 1857},
    {"titulo": "Quincas Borba", "autor": "Machado de Assis", "ano": 1891}
]

titulo = input("Digite o titulo do livro: ")

resultado = busca_binaria_livro(biblioteca, titulo)

if resultado:
    print(f"Titulo: {resultado['titulo']}")
    print(f"Autor: {resultado['autor']}")
    print(f"Ano: {resultado['ano']}")
else:
    print("Livro nao encontrado")
```
