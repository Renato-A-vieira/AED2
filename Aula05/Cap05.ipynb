```python
class Livro:
    def __init__(self, titulo, autor, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.__quantidade = quantidade
    
    def get_quantidade(self):
        return self.__quantidade
    
    def emprestar(self):
        if self.__quantidade > 0:
            self.__quantidade -= 1
            return True
        return False
    
    def devolver(self):
        self.__quantidade += 1
    
    def __str__(self):
        return f"{self.titulo} - {self.autor} ({self.__quantidade} disponiveis)"


class Usuario:
    def __init__(self, nome, matricula):
        self.nome = nome
        self.matricula = matricula
        self.livros_emprestados = []
    
    def emprestar_livro(self, livro):
        if livro.emprestar():
            self.livros_emprestados.append(livro)
            print(f" '{self.nome} emprestou '{livro.titulo}'")
            return True
        else:
            print(f" '{livro.titulo}' indisponivel para {self.nome}")
            return False
    
    def devolver_livro(self, livro):
        if livro in self.livros_emprestados:
            livro.devolver()
            self.livros_emprestados.remove(livro)
            print(f" {self.nome} devolveu '{livro.titulo}'")
            return True
        else:
            print(f" {self.nome} nao possui '{livro.titulo}'")
            return False
    
    def listar_emprestimos(self):
        if self.livros_emprestados:
            print(f"Livros emprestados para {self.nome}:")
            for livro in self.livros_emprestados:
                print(f"  - {livro.titulo}")
        else:
            print(f"{self.nome} nao possui livros emprestados")
    
    def __str__(self):
        livros = ", ".join([l.titulo for l in self.livros_emprestados]) if self.livros_emprestados else "nenhum"
        return f"{self.nome} (Matricula: {self.matricula}) | Livros: {livros}"


print("=" * 50)
print("SISTEMA DE BIBLIOTECA")
print("=" * 50)
print()

print("CADASTRO DE LIVROS")
livro1 = Livro("Codigo da vinci", "Dan Brown", 2)
livro2 = Livro("O Alquimista", "Paulo Coelho", 1)
livro3 = Livro("Dom Casmurro", "Machado de Assis", 0)

print(livro1)
print(livro2)
print(livro3)
print()

print("--- CADASTRO DE USUARIOS ---")
usuario1 = Usuario("Renato", "2025001")
usuario2 = Usuario("Amanda", "2025002")

print(usuario1)
print(usuario2)
print()

print("OPERACOES DE EMPRESTIMO")
usuario1.emprestar_livro(livro1)
usuario1.emprestar_livro(livro2)
usuario2.emprestar_livro(livro1)
usuario2.emprestar_livro(livro3)
print()

print("SITUACAO APOS EMPRESTIMOS")
print(livro1)
print(livro2)
print(livro3)
print()
usuario1.listar_emprestimos()
usuario2.listar_emprestimos()
print()

print("OPERACOES DE DEVOLUCAO")
usuario1.devolver_livro(livro1)
usuario2.devolver_livro(livro1)
print()

print("SITUACAO FINA")
print(livro1)
print(livro2)
print(livro3)
print()
print(usuario1)
print(usuario2)
```

```python
class Conta:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo
    
    def depositar(self, valor):
        if valor > 0:
            self.saldo += valor
            print(f"Deposito de R$ {valor:.2f} realizado. Novo saldo: R$ {self.saldo:.2f}")
        else:
            print("Valor invalido para deposito")
    
    def sacar(self, valor):
        if valor > 0 and valor <= self.saldo:
            self.saldo -= valor
            print(f"Saque de R$ {valor:.2f} realizado. Novo saldo: R$ {self.saldo:.2f}")
            return True
        elif valor > self.saldo:
            print(f"Saldo insuficiente. Saldo atual: R$ {self.saldo:.2f}")
            return False
        else:
            print("Valor invalido para saque")
            return False
    
    def __str__(self):
        return f"Titular: {self.titular} | Saldo: R$ {self.saldo:.2f}"


class ContaCorrente(Conta):
    def __init__(self, titular, saldo, taxa_saque=0.50, limite_especial=500):
        super().__init__(titular, saldo)
        self.taxa_saque = taxa_saque
        self.limite_especial = limite_especial
    
    def sacar(self, valor):
        valor_total = valor + self.taxa_saque
        saldo_disponivel = self.saldo + self.limite_especial
        
        if valor > 0 and valor_total <= saldo_disponivel:
            self.saldo -= valor_total
            print(f"Saque de R$ {valor:.2f} + taxa R$ {self.taxa_saque:.2f} realizado")
            print(f"Novo saldo: R$ {self.saldo:.2f}")
            return True
        elif valor_total > saldo_disponivel:
            print(f"Limite insuficiente. Saldo + limite: R$ {saldo_disponivel:.2f}")
            return False
        else:
            print("Valor invalido para saque")
            return False
    
    def __str__(self):
        return f"Conta Corrente | {super().__str__()} | Taxa: R$ {self.taxa_saque:.2f} | Limite: R$ {self.limite_especial:.2f}"


class ContaPoupanca(Conta):
    taxa_juros = 0.05
    
    def __init__(self, titular, saldo):
        super().__init__(titular, saldo)
    
    def render_juros(self):
        juros = self.saldo * self.taxa_juros
        self.saldo += juros
        print(f"Juros de R$ {juros:.2f} aplicados. Novo saldo: R$ {self.saldo:.2f}")
    
    def __str__(self):
        return f"Conta Poupanca | {super().__str__()} | Taxa juros: {self.taxa_juros * 100:.0f}%"


print("=" * 60)
print("SISTEMA BANCARIO")
print("=" * 60)
print()

print("--- CONTA CORRENTE ---")
cc = ContaCorrente("Fabio", 1000, taxa_saque=0.50, limite_especial=500)
print(cc)
cc.depositar(200)
cc.sacar(300)
cc.sacar(1500)
print(cc)
print()

print("--- CONTA POUPANCA ---")
cp = ContaPoupanca("Rayssa", 2000)
print(cp)
cp.depositar(500)
cp.sacar(300)
cp.render_juros()
print(cp)
print()

print("--- OPERACOES ADICIONAIS ---")
print("Tentando sacar mais que o saldo na poupanca:")
cp.sacar(3000)
print()
cp.render_juros()
print(cp)
print()

print("--- TESTE DE HERANCA ---")
contas = [cc, cp]
for conta in contas:
    print(conta)
```
