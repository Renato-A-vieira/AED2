```python
import random
lista_palavras =['banana','Palmito','Praia','Paralelepipedo','Hipopotamo','Carreta']
palavra_escolhida = random.choice(lista_palavras)


palco = ['','O','-','|','<', '_|>']
class Forca:
  def __init__(self,palavra):
    self.palavra = palavra.lower()
    self.letra_certa = []
    self.letra_errada = []

  def adivinhar_letra(self,letra):
    if letra in self.palavra and letra not in self.letra_certa:
      self.letra_certa.append(letra)
    elif letra not in self.palavra and letra not in self.letra_errada:
      self.letra_errada.append(letra)
    else:
      return False
    return True

  def forca_acabou(self):

    if self.forca_venceu() or (len(self.letra_errada) >= 5):
      return True
    return False

  def forca_venceu(self):

    return '_' not in self.palavra_escondida()

  def palavra_escondida(self):
    palavra_formatada = ''
    for letra in self.palavra:
      if letra not in self.letra_certa:
        palavra_formatada += '_'
      else:
        palavra_formatada += letra
    return palavra_formatada

  def mostra_status(self):
    print('\n +++++++ Jogo da Forca +++++++')
       print(palco[len(self.letra_errada)])
    print(f'\n Palavra: {self.palavra_escondida()}')
    print('\n Letras Erradas: ')
    for letra in self.letra_errada:
      print(letra, end=' ')
    print('\n Letras Certas: ')
    for letra in self.letra_certa:
      print(letra, end=' ')
    print()

def main():
  jogo = Forca(palavra_escolhida)
  while not jogo.forca_acabou():
      jogo.mostra_status()
      letra_escolhida = input('\n Escolha uma letra: ').lower()
      jogo.adivinhar_letra(letra_escolhida)


  jogo.mostra_status()

  if jogo.forca_venceu():
    print('\n Parabéns, você venceu!')
  else:
    print('\n Você perdeu!')
    print('A palavra era ' + jogo.palavra)

print('Foi bom jogar com você')
main()
```
