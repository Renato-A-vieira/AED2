```python
class Fila:
    def __init__(self):
        self.itens = []
    def enqueue(self, item):
        self.itens.append(item)
    def dequeue(self):
        return self.itens.pop(0)
    def size(self):
        return len(self.itens)
    def is_empty(self):
        return len(self.itens) == 0
```

```python
class Queue:
    def __init__(self):
        self.itens = []
    def enqueue(self, item):
        self.itens.append(item)
    def dequeue(self):
        return self.itens.pop(0)
    def size(self):
        return len(self.itens)


def batata_quente(criancas, passes):
    fila = Queue()
    for nome in criancas:
        fila.enqueue(nome)
    while fila.size() > 1:      
        for i in range(passes):
            fila.enqueue(fila.dequeue())      
        eliminado = fila.dequeue()
        print(f'Saiu: {eliminado}')
    return fila.dequeue()
    
a = batata_quente(['Renato','Julia','Amanda','Kaio','Ana','Harry', 'ze'], 55)
print(f'O vencedor(a) é {a}')
```
